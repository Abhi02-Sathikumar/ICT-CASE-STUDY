```python
#14) Group by Director: avg Rating, Revenue
movie.groupby("Director")[["Rating","Revenue (Millions)"]].mean("Rating")
```

```python
#15) Group by Year: count, avg Rating
movie.groupby("Year")[["Rating"]].agg(["count","mean"])
```

```python
#16) Create a new column Profitability:
#a) Revenue divided by Duration
movie["Profitability"] = movie["Revenue (Millions)"]/movie["Runtime (Minutes)"]
movie
```

```python
#b) Find top 10 most profitable movies
movie.sort_values("Profitability",ascending=False).head(10)
```

```python
#17) Create a column that categorizes movies by rating:
#a) Rating < 6: Poor
#b) 6 ≤ Rating < 7.5: Average
#c) Rating ≥ 7.5: Excellent
def rating_categories(Rating):
  if Rating <6:
    return "poor"
  elif 6 <= Rating < 7.5:
    return "average"
  else:
    return "excellent"

movie["rating_category"] = movie["Rating"].apply(rating_categories)
movie
```

```python
#18) Create a custom function to score movies based on:
#i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
#Apply a function to dataset to assign the score
def score(row):
  high_score = row['Rating'] * 10 + row['Revenue (Millions)']
  return high_score

movie["scores"] = movie.apply(score,axis =1)
movie
```
