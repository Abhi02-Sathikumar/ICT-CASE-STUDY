```python
import pandas as pd
import sqlite3

def extract_data():
    """Extracts data from various sources."""
    print("Extracting data...")
    population_data = pd.read_csv('/content/population.csv')
    gdp_data = pd.read_excel("/content/gdp.xlsx")
    internet_df = pd.read_json('/content/internet_users.json')
    literacy_df = pd.read_xml('/content/literacy_rate.xml')
    print("Data extracted successfully.")
    return population_data, gdp_data, internet_df, literacy_df

population_data, gdp_data, internet_df, literacy_df = extract_data()
```

```python
def transform_data(population_data, gdp_data, internet_df, literacy_df):
    """Cleans, renames, merges, and creates new features."""
    print("Transforming data...")

    gdp_data.rename(columns={gdp_data.columns[0]:'country'},inplace=True)
    literacy_df.rename(columns={literacy_df.columns[0]:'country'},inplace=True)

    population_data['population'] = pd.to_numeric(population_data['population'], errors='coerce')
    gdp_data['GDP'] = pd.to_numeric(gdp_data['GDP'], errors='coerce')
    internet_df['internet_users'] = pd.to_numeric(internet_df['internet_users'] ,errors='coerce')
    literacy_df['literacy_rate'] = pd.to_numeric(literacy_df['literacy_rate'], errors='coerce')

    population_data['country'] = population_data['country'].astype(str)
    gdp_data['country'] = gdp_data['country'].astype(str)
    internet_df['country'] = internet_df['country'].astype(str)
    literacy_df['country'] = literacy_df['country'].astype(str)

    merged_1 = pd.merge(population_data, gdp_data, on='country', how='inner')
    merged_2 = pd.merge(merged_1, internet_df, on='country', how='inner')
    final_merge = pd.merge(merged_2, literacy_df, on='country', how='inner')

    final_merge['internet_penetration'] = (final_merge['internet_users'] / final_merge['population']) * 100
    final_merge['internet_penetration'] = final_merge['internet_penetration'].round(2)
    final_merge['gdp_per_capita'] = final_merge['GDP'] / final_merge['population']
    final_merge['gdp_per_capita'] = final_merge['gdp_per_capita'].round(2)

    print("Data transformed successfully.")
    return final_merge

final_merge = transform_data(population_data, gdp_data, internet_df, literacy_df)
display(final_merge.head())
```

```python
def load_data(final_merge, db_name='country_statistics.db', csv_name='final_country_statistics_etl.csv'):
    print(f"Loading data to {db_name} and {csv_name}...")

    conn = sqlite3.connect(db_name)
    final_merge.to_sql('country_statistics', conn, if_exists='replace', index=False)
    conn.close()

    final_merge.to_csv(csv_name, index=False)

    print("Data loaded successfully.")


load_data(final_merge)
```
